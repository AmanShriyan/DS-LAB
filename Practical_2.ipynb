```python
missing_values = ["Na", "NaN"]
df = pd.read_csv("StudentsPerformance.csv", na_values = missing_values)
df
```

```python
df["reading score"] = df["reading score"].fillna(df["reading score"].mean())
df
```

```python
df.replace(to_replace = 0, value = -99)
df
```

```python
df.dropna(how = "all")
```

```python
df.dropna(axis = 1)
```
