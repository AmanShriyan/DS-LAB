```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
import numpy as np

dt = sns.load_dataset('titanic')
dt.head()
```

```python
# Distplot visualization

# displt = sns.distplot(x=data['fare'], bins=10)

displt = sns.distplot(x=dt['age'], bins=20, color='red', kde=False)

# displt
```

```python
# Joint Plot Visualization

jtpt = sns.jointplot( x = dt['age'], y = dt['sex'], kind='scatter')
```

```python
bar = sns.barplot(x = 'sex', y = 'age', data=dt)

# bar = sns.barplot(x = 'sex', y = 'age', data=dt, estimator=np.std)

```

```python
ctplt = sns.countplot(x = 'age', data=dt)
```

```python
# Box plot

bxplt = sns.boxplot(x = 'sex', y= 'age', data=dt, hue='survived')
```

```python
# Bar plot

bar = sns.barplot(x = 'sex', y = 'age', data=dt, estimator=np.std)
bar
```

```python
vplt = sns.violinplot(x = 'sex', y = 'age', data=dt)
```

```python
# swplt = sns.swarmplot( x = 'sex', y = 'age', data=dt)
swplt = sns.swarmplot( x = 'sex', y = 'age', data=dt, hue='survived')

```

```python
htplt = sns.histplot(dt['age'], kde=False, bins = 10)
```
